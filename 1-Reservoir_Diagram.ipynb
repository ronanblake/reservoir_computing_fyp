## Network visualisation

```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt

N = 20 
p = 0.055
G = nx.DiGraph()
G.add_nodes_from(range(N))
rng = np.random.default_rng(8)

for i in range(N):
    for j in range(N):
        if rng.random() < p: #if a random number between 0,1 is less than p, then add an edge between node i and node j
            G.add_edge(i, j)   # (i,i) gives a self-loop
G.add_edge(5, 5) #ensuring at least one self loop
pos = nx.spring_layout(G, seed=2)

plt.figure(figsize=(6,6))
nx.draw(G, pos, node_size=500, width = 3,node_color="black", arrows=True)
plt.savefig("reservoir_network.svg", transparent=True) 
plt.show()
plt.close()
```
